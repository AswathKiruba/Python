```python
#import workbook
from xlsxwriter import Workbook
#import requests
import requests
#import Beautiful Soups
from bs4 import BeautifulSoup
#import userAgent
from fake_useragent import UserAgent
#import date time
import datetime as dt
import httplib
from urlparse import urlparse
import pandas as pd
```

```python
#creating a user Agent
ua=UserAgent()
header ={'user-agent':ua.chrome}
page= requests.get('http://www.manutd.com//',headers=header)
print page
#print page.status_code
```

```python
def FindActiveLinks(webpage):
    soup = BeautifulSoup(webpage.content,'lxml')
    List1 = []
    List1.append(['Anchor tag','Link','Text','Valid','DateTime'])
    for link in soup.findAll('a'):
        list2 = []
        list2.append(link)
        href = link.get('href')
        list2.append(href)
        s = link.string
        if s== None:
            s = str(s)    
        v = s.encode('utf-8')
        list2.append(v)
        def uri_validator(href):
            try:
                result = urlparse(href)
                #print 'valid'
                list2.append(True)
                list2.append(str(dt.datetime.now()))
                return True if [result.scheme, result.netloc, result.path] else False
            except:
                #print 'not valid'
                
                list2.append(False)
                list2.append(str(dt.datetime.now()))
                return False
        uri_validator(href)
        List1.append(list2)

    return List1   
    
```

```python
FindActiveLinks(page)
```

```python
# writing to a excel file
csv = pd.DataFrame(FindActiveLinks(page))
csv.to_csv('Manchester.csv',index=False,header=False)
```
