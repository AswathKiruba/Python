## Searching elements and using text fields

```python
  #make a webdriver object 
driver = webdriver.Chrome('C:/Users/aswat/Downloads/chromedriver_win32/chromedriver')

#get the web page
driver.get('https://www.google.com/')

soup = BeautifulSoup(driver.page_source,'lxml')

#search a tag using id
searchbar = driver.find_element_by_id('lst-ib')

# input data
searchbar.send_keys('Fidelity investments')

#submit form
searchbar.submit()

sleep(10)

#close webrdiver object
driver.close()
```

## Clicking element or href

```python
  #make a webdriver object 
driver = webdriver.Chrome('C:/Users/aswat/Downloads/chromedriver_win32/chromedriver')

#get the web page
driver.get('https://www.instagram.com/?hl=en')

soup = BeautifulSoup(driver.page_source,'lxml')

#find element by link
login_button = driver.find_element_by_link_text('Log in')

#click element
login_button.click()


sleep(10)

#close webrdiver object
driver.close()
```

## XPath

```python
# instagram example
driver = webdriver.Chrome('C:/Users/aswat/Downloads/chromedriver_win32/chromedriver')

driver.get('https://www.instagram.com')

sleep(2)
login_button = driver.find_element_by_xpath("//span[@id='react-root']//p[@class='_c59vy']/a") # valid

login_button.click()

sleep(5)

#close webrdiver object
driver.close()
```
